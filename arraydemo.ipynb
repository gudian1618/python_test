```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab


mu = 100
sigma = 15
x = mu + sigma * np.random.randn(10000)
print("x:",x.shape)
%matplotlib inline
num_bins = 50
n, bins, patches = plt.hist(x, num_bins, normed=1, alpha=0.8, histtype='bar')
y = mlab.normpdf(bins, mu, sigma)
plt.plot(bins, y, 'r--')
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ: $\mu=100$, $\sigma=15$')
plt.subplots_adjust(left=0.15)
plt.show()
print('bind:\n',bins)
```

```python
import numpy as np
from matplotlib import cm
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
% matplotlib qt5
# 生成数据
delta = 0.2 
x = np.arange(-3, 3, delta)
y = np.arange(-3, 3, delta)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
x = X.flatten()
y = Y.flatten()
z = Z.flatten()
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_trisurf(x,y,z, cmap=cm.jet, linewidth=0.01)
plt.title('3D')
ax2 = fig.add_subplot(122)
cs = ax2.contour(X, Y, Z, 15, cmap='jet')
ax2.clabel(cs, inline=True, foontsize=10, fmt='%1.1f')
plt.title('countour')
plt.show()
```
